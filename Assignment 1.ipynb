Question:2. Python Program to Add, Subtract, Multiply and Divide two numbers.


````python
#input numbers
print("\033[1mInput Numbers:\033[0m\n")
num1=int(input("First Number: "))
num2=int(input("Second Number:"))

print("\n\033[1mDesire Outputs:\033[0m")
#Add
add = num1+num2
print("Addition: ",add)

# Subtraction
sub = num1-num2
print("Subtraction: ",sub)

# Multiplication
Multi = num1*num2
print("Multiplication: ",Multi)

# Division
Div = num1/num2
print("Division: ",Div)
````


Question:3. Check if the first and last number of a list is the same.


```
#
Input: [5,6,7,8,5]
Output: True

Input: [5,6,7,8,9]
Output: False
```


````python
#List input
list=[]
num = int(input("Enter the number of elements:"))

for i in range(0,num):
  elements = int(input())
  list.append(elements)
print("input elements are: ",list)
````

````python
#List_1 input
list1=[]
num1 = int(input("Enter the number of elements for list_1:"))

for i in range(0,num1):
  elements_1 = int(input())
  list1.append(elements_1)
print("input elements for list_1:",list1)

# to get first element & last element of list_1
first_number_1= [list1[0]]
last_number_1 = [list1[-1]]
print("\nFirst number of list_1",first_number_1)
print("last number of list_1",last_number_1)

#Check if the first and last number of a list is the same.
if first_number_1 == last_number_1:
  print("\n\033[1mOutput:\033[0m",True)
else:
  print("\n\033[1mOutput:\033[0m",False)

#List_2 input
list2=[]
num2 = int(input("\nEnter the number of elements for list_2:"))
for i in range(0,num2):
  elements_2 = int(input())
  list2.append(elements_2)
print("\ninput elements for list_2:",list2)

# to get first element & last element of list_2
first_number_2= [list2[0]]
last_number_2 = [list2[-1]]
print("\nFirst number of list_2",first_number_2)
print("last number of list_2",last_number_2)

#Check if the first and last number of a list is the same.
if first_number_2 == last_number_2:
  print("\n\033[1mOutput:\033[0m",True)
else:
  print("\n\033[1mOutput:\033[0m",False)
````
