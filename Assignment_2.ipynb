1. Write a Python program Return multiple values from a function
```
# Input: 100 50
Output:
Addition: 150
subtraction: 50
```




````python
print("\033[1mInput:\033[0m")
num1=int(input())
num2=int(input())

def add(num1,num2):
  sum = num1+num2
  return sum
def sub(num1,num2):
  subs = num1-num2
  return subs

addition = add(num1,num2)
subratction = sub(num1,num2)

print("\033[1moutput:\033[0m")
print("Addition",addition)
print("Subratction",subratction)

````

3. Find the largest item from the given three numbers using a function.
```
# Input: 40,37,53
Output: 53
```


````python
list=[]
n = int(input("Enter number of elements : "))
for i in range(0, n):
  elements = int(input())
  list.append(elements)
print("Output: ",list)
````

````python
list=[]
n = int(input("Enter number of elements : "))
for i in range(0, n):
  elements = int(input())
  list.append(elements)
print("Input: ",list)

def largest(list,n):
  list.sort()
  return list[n-1]

n = len(list)
Ans = largest(list, n)
print("Output:\nLargest in given numbers ", Ans)
````
